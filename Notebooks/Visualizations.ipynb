```python
import dash
import dash_core_components as dcc
import dash_html_components as html
import plotly.express as px
import pandas as pd

# Sample data for demonstration

# Create the scatter plot
fig = px.scatter(df, x='Time_taken (min)', y='Distance in kms', color='Weather_conditions',
                 title='Time_taken (min) vs Distance in kms',
                 labels={'Time_taken (min)': 'Time_taken (min)', 'Distance in kms': 'Distance in kms'},
                 color_discrete_sequence=px.colors.qualitative.Set1)

# Customize the layout for better visuals
fig.update_layout(
    title_font_size=24,
    paper_bgcolor='rgb(230, 230, 250)',  # Light lavender background
    plot_bgcolor='rgb(240, 240, 240)',   # Light grey background for the plot area
    legend_title_text='Weather Conditions',
    legend=dict(
        x=0.8, y=0.99,
        bgcolor='rgba(255, 255, 255, 0.5)',
        bordercolor='Black',
        borderwidth=2
    ),
    xaxis=dict(
        title='Time_taken (min)',
        gridcolor='white'
    ),
    yaxis=dict(
        title='Distance in kms',
        gridcolor='white'
    )
)

# Increase marker size for better visibility
fig.update_traces(marker=dict(size=12, line=dict(width=2, color='DarkSlateGrey')))

# Initialize the Dash app
app = dash.Dash(__name__)

# Define the layout of the app
app.layout = html.Div(style={'backgroundColor': 'rgb(230, 230, 250)', 'padding': '10px'}, children=[
    html.H1('Time vs Distance Scatter Plot', style={'textAlign': 'center', 'color': 'DarkSlateGrey'}),
    dcc.Graph(
        id='scatter-plot',
        figure=fig
    )
])

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```

# Frequency encoding

```python
temp=df['Delivery_person_ID'].value_counts().reset_index()
h={}
h=dict(h)
for i in range(len(temp['Delivery_person_ID'])):
    h[str(temp['Delivery_person_ID'][i])]=temp['count'][i]

df['Delivery_person_ID']=df['Delivery_person_ID'].replace(h)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Assuming df is your DataFrame
corr = np.round(df.select_dtypes(exclude='object').corr(), 2)

# Create a figure with specific size
plt.figure(figsize=(10, 8))  # Adjust width and height as needed

# Plot the heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Convert hours to radians
hours = df['Time_Order_picked_Hour']
radians = hours * (2 * np.pi / 24)

# Create a polar plot
plt.figure(figsize=(8, 6))
ax = plt.subplot(111, projection='polar')
ax.hist(radians, bins=24, edgecolor='k', alpha=0.75)

ax.set_theta_zero_location('N')
ax.set_theta_direction(-1)
ax.set_xticks(np.linspace(0, 2 * np.pi, 24, endpoint=False))
ax.set_xticklabels(range(24))

plt.show()
```
